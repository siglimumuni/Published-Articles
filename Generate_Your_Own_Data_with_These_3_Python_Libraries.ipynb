#Faker


```python
#import module
from faker import Faker

#create an instance of Faker()
fake = Faker()

#generate data by accessing different properties
print(fake.name())
print(fake.address())
print(fake.date_of_birth(minimum_age = 18))
print(fake.job())
print(fake.city())
print(fake.color_name())
```

```python
#specify locales
fake = Faker(['it_IT', 'de_DE', 'zh_CN'])
for _ in range(10):
    print(fake.name())
```

```python
import pandas as pd

#set locale
fake = Faker('fr_FR')

#create an empty list for each column
name, address, phone_number, dob, job = [[] for k in range(5)]

#create 200 rows of data by appending properties to the above lists
for row in range(200):
  name.append(fake.name())
  address.append(fake.address())
  phone_number.append(fake.phone_number())
  dob.append(fake.date_of_birth(minimum_age=25))
  job.append(fake.job())

#create a dictionary and map column names to corresponding lists
data = {"Full Name":name, "Address":address, "Phone Number":phone_number, "Date of Birth":dob, "Job Title":job}

#create dataframe from dictionary
df = pd.DataFrame(data)
df.head()
```

```python
#create dataframe from profiles
fake = Faker()
df = pd.DataFrame()
for i in range(1000):
    df = df.append(fake.profile(), ignore_index=True)

df.head()
```

# Sk-Learn

##Make blobs

```python
from sklearn.datasets import make_blobs
import pandas as pd
# generate classification dataset with 2 features and 3 labels
X, y = make_blobs(n_samples=200, centers=3, n_features=2)

#create dataframe
df = pd.DataFrame(dict(feature_1=X[:,0], feature_2=X[:,1], label=y))

df.head(6)
```

```python
import matplotlib.pyplot as plt

#create scatter plots
fig, ax = plt.subplots()
colors = {0:'blue', 1:'red', 2:'green'}
groups = df.groupby('label')
for i, group in groups:
    group.plot(ax=ax, kind='scatter', x='feature_1', y='feature_2', label=i, color=colors[i])
plt.show()
```

##Make moons

```python
from sklearn.datasets import make_moons
import pandas as pd

# generate binary classification dataset
X, y = make_moons(n_samples=200, noise=0.1)

# create dataframe
df = pd.DataFrame(dict(feature_1=X[:,0], feature_2=X[:,1], label=y))
df.head(6)
```

```python
import matplotlib.pyplot as plt

# create scatter plots
fig, ax = plt.subplots()
colors = {0:'blue', 1:'red'}
groups = df.groupby('label')
for i, group in groups:
    group.plot(ax=ax, kind='scatter', x='feature_1', y='feature_2', label=i, color=colors[i])
plt.show()
```
