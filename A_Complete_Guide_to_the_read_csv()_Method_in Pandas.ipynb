```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/CRM_Data.csv")

df.head()
```

## 1. Promoting rows as headers

```python
#with the skip rows argument
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/CRM_Data.csv",header=2)

df.head()
```

## 2. Parsing dates in a specified column

```python
#read the data
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/AirQuality")
#check the datatypes
print(df.dtypes)
```

```python
#read the data with parse_dates argument
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/AirQuality",
                 parse_dates=['Timestamp'])
#check the datatypes
print(df.dtypes)
```

## 3. Setting a column as an index

```python
#read the data and set timestamp column as index
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/AirQuality",
                 parse_dates=['Timestamp'],
                 index_col='Timestamp')

df.head()
```

## 4. Setting the data type of a column


```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/Ecommerce%20Customers")
df.dtypes
```

```python
#change column data types
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/Ecommerce%20Customers",
                 dtype={'Email':'string','Address':'string','Avatar':'string'})
df.dtypes
```

## 5. Setting boolean values


```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/churn.csv")
df.head()
```

```python
#change boolean values
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/churn.csv",
                 true_values = ['yes'],
                 false_values = ['no'])
df.head()
```

## 6. Assigning null values

```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/sample_data.csv")
df.head()
```

```python
#return count of null values
df.isnull().sum()
```

```python
#identify null values
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/sample_data.csv",
                 na_values = ['?'])
df.head()
```

```python
#return count of null values
df.isnull().sum()
```

## 7. Reading a fraction of the rows in the data


```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/yelp.csv")
df.shape
```

```python
#select first 1000 rows
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/yelp.csv",
                 nrows=1000)
df.shape
```

```python
#skip first 7000 rows
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/yelp.csv",
                 skiprows=7000)
df.shape
```

## 8. Selecting specific columns from the data

```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/yelp.csv")
df.head()
```

```python
#keep three columns with list of column names
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/yelp.csv",
                 usecols = ['date','stars','text'])
df.head()
```

```python
#keep three columns with list of column index positions
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/yelp.csv",
                 usecols = [1,3,4])
df.head()
```
