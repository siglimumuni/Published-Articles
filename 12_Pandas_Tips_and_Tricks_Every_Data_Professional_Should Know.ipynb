## 1. Specifying the maximum number of rows or columns to display

```python
import pandas as pd

#set the max number of rows 
pd.set_option("max_rows",200)

#check the max number of rows
pd.options.display.max_rows
```

```python
#set the max number of columns
pd.set_option("max_columns",40)

#check the max number of columns
pd.options.display.max_columns
```

## 2. Subsetting a dataframe with the query() method

```python
#read dataset
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/advertising.csv")
df.head()
```

```python
#show only rows with the "age" column above 30
df_subset = df[(df['Age'] > 30) & (df['Country']=='Australia')]
df_subset.head()
```

```python
#using the query method
df_subset = df.query("Age > 30 & Country == 'Australia'")
df_subset.head()
```

## 3. Changing the number of decimal places in a column with the round() method

```python
#change values in column to one decimal place
df['Daily Internet Usage'] = df['Daily Internet Usage'].round(1)
df['Daily Internet Usage'].head()
```

## 4. Unpacking list values with the explode() method

```python
df = pd.DataFrame({"Name":["Mike","Susan","Joel"],"Favorite Fruit":["Mango","Orange",["Pineapple","Orange","Strawberry"]]})
df
```

```python
#unpack list values into separate observations
df.explode("Favorite Fruit")
```

## 8. Randomly sample rows with sample()

```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/advertising.csv")

df.shape
```

```python
#extract 100 samples
df1 = df.sample(n=100)
df1.shape
```

```python
#extract a quarter of the dataframe
df2 = df.sample(frac=0.25)
df2.shape
```

## 9. Return a cumulative summary with cumsum()

```python
df = pd.DataFrame({'Date':['2022-08-01','2022-08-02','2022-08-03','2022-08-04','2022-08-05','2022-08-06'],
                   'Total Sales':[2000,3000,2500,4000,3000,2000]
                   })
df
```

```python
#create new column with cum summary of total sales
df['Cummulative Sales'] = df['Total Sales'].cumsum()

df
```

## 10. Aggregate by multiple functions using agg()

```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/Mall_Customers.csv")
df.head()
```

```python
#group by gender and return mean of every column
df_mean = df.groupby('Gender').mean()
df_mean
```

```python
#aggregate each column by a different function
df_agg = df.groupby('Gender').agg({'CustomerID':'count',
                                   'Age':'mean',
                                   'Annual Income (k$)':'max',
                                   'Spending Score (1-100)':'min'})
df_agg
```

## 11. Aggregating time series data with the resample() method

```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/AirQuality",
                 parse_dates=['Timestamp'],
                 index_col='Timestamp')

df.head()
```

```python
#resample data by avg nitroux oxide measurement per month
df['NO2(GT)'].resample('M').mean()
```

## 12. Performing multi-variable counts with crosstab()






```python
df = pd.read_csv("https://raw.githubusercontent.com/siglimumuni/Datasets/master/bank.csv")

df.head()
```

```python
#return count by marital status and education
pd.crosstab(df['marital'],df['education'])
```

```python
#add row and column subtotals
pd.crosstab(df['marital'],df['education'], margins=True)
```

```python
#return percentages instead of counts
pd.crosstab(df['marital'],df['education'], normalize=True)
```
